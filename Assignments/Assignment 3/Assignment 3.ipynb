### Question:  
1.	Load both CSVs.
2.	Store loads and prices into lists.  
3.	Write a function to compute.  
3.1.	Hourly bill  
3.2.	Total bill


```python
import csv          #importing csv module
load=[]
with open(r"D:\3rd semester\introduction to AI\Assignments and OEL\Assignment 3 load.csv") as f1:  #1.reading given load csv file.
    reader1=csv.reader(f1)
    for index,r1 in enumerate(reader1):    #2.storing loads in list by for loop
        if index==0:    #skips the 1st row 
         continue
        load.append(float(r1[1]))
load  
```

```python
price=[]
with open('Assignment 3 price.csv') as f2:              #1.reading given price csv file.
    reader2=csv.reader(f2)
    for index,r2 in enumerate(reader2):             #2.storing price in list by for loop
        if index==0:       #skips the 1st row 
         continue
        price.append(float(r2[1]))
price
```

```python
unit=[x/1000 for x in load]         #converting to kwh 1kwh=1unit for bill calculation.
unit
```

```python

def hourly_bill(unit,price):            #3.creating function to calculate bill.
    bill=[]
    for k in range(len(unit)):
        bill.append(unit[k]*price[k])  #3.1 calculating hourly bill
    return bill
bill=hourly_bill(unit,price)            #calling function and providing with price and units(kwh)
for i,b in enumerate(bill):
      print(f"Hour no: {i+1} bill is: {b}")
print("\n Total bill is:",sum(bill))        #3.2 calculating and printing total bill. 

 
 
```
