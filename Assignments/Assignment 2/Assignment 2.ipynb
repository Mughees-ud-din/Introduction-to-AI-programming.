2. Group consumption by hour (manual group-by)
Create a dictionary like:  
{ 1: [200, 700, 800, 500],  
  2: [300, 600, 700, 400],  
  3: [500, 400, 600, 650] }    
2.1.	Compute average consumption per hour  
2.2.	Compute peak consumption per hour  
2.3.	Print results in a neat format



```python
group={ 1: [200, 700, 800, 500],2: [300, 600, 700, 400],3: [500, 400, 600, 650] }
avg=dict()
peak=dict()
for i,g in enumerate(group.values()):
    avg[i+1]=sum(g)/len(g)      #2.1 compting average consumption per hour
    peak[i+1]=max(g)        #2.1 compting peak consumption per hour 
```

```python
#2.3 printing result
for i,a in avg.items():
    print(f"Hour: {i} avg consumption is->",a)
print("\n")
for i,p in peak.items():
    print(f"Hour: {i} peak consumption is->",p)
```

3.	Group consumption by day  
3.1.	Create a dictionary like:  
{ 7: [200,300,500,700,600,400],  
  8: [800,700,600,500,400,650]  }  
3.2.	Compute total consumption per day  
3.3.	Compute daily average.


```python
group={7:[200,300,500,700,600,400],8:[800,700,600,500,400,650]}
avg=dict()
sam=dict()
for i,g in enumerate(group.values()):
    sam[i+1]=sum(g)             #3.2 calculating total consumption per day
    avg[i+1]=sum(g)/len(g)      #3.3 calculating avg consumption per day
```

```python

for i,a in avg.items():
    print(f"Day: {i} avg consumption is->",a)
print("\n")
for i,p in sam.items():
    print(f"Day: {i} sum consumption is->",p)
```
