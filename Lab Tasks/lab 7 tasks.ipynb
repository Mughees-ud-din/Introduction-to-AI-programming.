## task 4:
Inventory System for Electronics Shop  
A shop tracks an item in its inventory:  
{  
    "item_id": 1042,  
    "name": "PWM Solar Charge Controller",  
    "stock": 18,  
    "price": 2500,  
    "features": {"voltage": "12/24V", "current": "20A", "brand": "SolarTech"}  
}  
1.Print the brand of the controller.  
2.Reduce the stock by 2 and update the dictionary.  
3.Add a new feature "warranty": "1 year".  
4.Check if "price" exists in the dictionary.  
5.Merge another dictionary {"discount": 10} into the inventory item.



```python
shop_tracks={
    "item_id": 1042,
    "name": "PWM Solar Charge Controller",
    "stock": 18,
    "price": 2500,
    "features": {"voltage": "12/24V", "current": "20A", "brand": "SolarTech"}
}
```

```python
print("brand of the controller is: ",shop_tracks["name"])
```

```python
shop_tracks["stock"]=shop_tracks["stock"]-2
print(shop_tracks)
```

```python
shop_tracks["warranty"]="1 year"
```

```python
"price" in shop_tracks
```

```python
another={"discount":10}
shop_tracks.update(another)
print(shop_tracks)
```

{

    "name": "Faisal",

2.Add a new medication "antibiotic": "6PM".  
3.Update heart rate to 78.

```python
patient_record={
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {"aspirin": "8AM", "vitamin_D": "12PM"}
}
```

```python
#1.Print the oxygen level.
print("oxygen level of patient is :",patient_record["oxygen_level"])
```

```python
#2.add a new medication
patient_record["medications"]["antibiotic"]="6PM"
```

```python
patient_record["heart_rate"]=78
```

```python
#4.Get the medication schedule for "insulin" using .get() with default "Not Prescribed". 
patient_record["medications"].get("insulin","not_prescribed")
```

```python
#5.Use .items() to print all key-value pairs.
print("key value pairs are :\n",patient_record.items())
```
