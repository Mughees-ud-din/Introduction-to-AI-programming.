You are building a program to analyze five resistor values in a circuit.  
•	Function 1: resistance():  return a list of 5 resistances.  
•	Function 2: voltage():  return voltages across each resistor.  
•	Function 3: current():  take both lists (Resistance & Voltage) and calculate currents.  
•	Display all lists nicely in the main program.  
•	Try two versions:  
o	Return as a tuple.  
o	Return as a dictionary with keys like "current_1".  


```python
#lab n0 4
def resistance():
    return [10, 20, 30, 40, 50]

def voltage_list():
    return [5, 10, 15, 20, 25]

def current_dictionary(R,V) -> dict:
    return {f"current_{i+1}": V[i]/R[i] for i in range(len(R))}
def current_touple(R,V):
    list1=[]
    for i in range(len(R)):
     list1.append(V[i]/R[i])
    return tuple(list1)  
R = resistance()
V = voltage_list()
I = current_dictionary(R, V)
I2 = current_touple(R,V)
print("Resistances:", R)
print("Voltages:", V)
print("Currents:", I)
print("Currents:", I2)
```

## task:6	Parallel Resistance:  
Now, calculate parallel resistance.  
•	Write a function parallel() that takes a list or tuple of resistances.  
•	Return the equivalent resistance.  
•	Add docstrings and annotations.  
•	Print the function’s __annotations__.


```python
#lab task 06
def parallel(resistances: list[float] | tuple[float]) -> float:
    """Calculate equivalent parallel resistance."""
    return 1 / sum(1/r for r in resistances)

print("Parallel Resistance:", parallel([10, 20, 30]))
print(parallel.__annotations__)
```

## task no:10	Solar Inverter Data (Dictionary):
A better inverter sends data as:  
{ "SOC": 60, "solar_generation": 2, "price": 60 }  
•	Define a function that receives this dictionary.  
•	Print the values in a clear format.  
•	Use dictionary unpacking.  
•	Add docstrings and annotations.  
•	Print __annotations__.


```python
#lab task no 10
def solar_dict(data: dict) -> None:
    """Print solar inverter data from dictionary"""
    for k, v in data.items():
        print(f"{k}: {v}")

solar_dict({"SOC": 60, "solar_generation": 2, "price": 60})
print(solar_dict.__annotations__)
```
