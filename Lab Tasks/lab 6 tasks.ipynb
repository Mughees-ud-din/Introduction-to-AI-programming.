## Task6: List comprehensions & conversion  
You have a tuple: values = (2, 4, 6, 8, 10).  
Convert the tuple to a list.  
Then use a list comprehension to produce a list where each value is squared.  
Filter this to include squared values greater than 30.  
Print the resulting list.

## Task 7: Operators & built-in functions with tuples  
Let nums = (3, 6, 9, 12)  
Use in and not in to test membership.  
Concatenate nums + (15, 18).  
Multiply nums * 3.  
Use len(), min(), max(), sum() on the tuple.


```python
nums = (3, 6, 9, 12)
#in
print(6 in nums)
print(10 in nums)
#not in
print(6 not in nums)
print(10 not in nums)
#concatenate
print(nums+(15,18))
#multiply
print(nums*3)
#len,min,max
print("length=",len(nums),"minimum=",min(nums),"maximum=",max(nums))
#sum
sum(nums)
```
