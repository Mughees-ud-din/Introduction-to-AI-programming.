## Task 1: Filter Voltage Readings   
You have a list of voltage readings from a sensor:  
voltages = [224, 198, 231, 177, 205, 189, 240]  
Create a list comprehension to extract only the readings below 200 volts.


```python
voltages = [224, 198, 231, 177, 205, 189, 240]
voltages=[v for v in voltages if v>200]
print(voltages)
```

## Task 2: Convert Power Ratings to kW
Given power values in watts:  
powers_w = [50, 120, 300, 750, 1500]  
Use a list comprehension to convert all values to kilowatts.


```python
powers_w = [50, 120, 300, 750, 1500]
powers_w=[x/1000 for x in powers_w ]
powers_w
```

## Task 3: Square Only Even Resistances  
You measured resistance values:  
resistors = [10, 15, 22, 47, 50, 68]  
Create a list of squares of only the even resistor values.


```python
resistors = [10, 15, 22, 47, 50, 68]
square=[r**2 for r in resistors if r%2==0]
square
```

## Task 5: Extract Appliance Names Starting With ‘F’  
Given appliance names:  
appliances = ["Fan", "Fridge", "Light", "Freezer", "Pump"]  
Use a list comprehension to extract only names starting with “F”.


```python
appliances = ["Fan", "Fridge", "Light", "Freezer", "Pump"]
appliances=[x for x in appliances if x.startswith("F")]
appliances
```
