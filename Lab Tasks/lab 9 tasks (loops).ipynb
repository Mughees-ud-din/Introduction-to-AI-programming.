## Task 1:
A lab keeps records of resistor values. Students need to print each resistance value for verification.  
Print each resistance value from a list of resistors.  
	Create a list, then use for loop to display each item.


```python
resistors=[12,13,14,15,20,25,30]
print("Resistors values are following :")
for r in resistors:
    print(r)
```

## Task 2:
Current readings are stored in a list. Engineers want to calculate the square of each value for power loss calculation.  
Print the square of each current reading using formula I².  
	Make a list of current values, loop and multiply each by itself. 


```python
current=[20,30,40,55,99,10,70,66]
sqrt=[]
for c in current:
    sqrt.append(c*c)
print("squared currents are :",sqrt)
```

## Task 4:
A power company records voltage values at different times of the day. Students must print all these values for analysis.  
Print each voltage value from a list of voltages.


```python
voltages=[220,330,440,550,1000,11000]
for i,v in enumerate(voltages):
    print(f"voltages {i}: {v}V")
```

## Task 5:
 A lab has capacitor values stored in microfarads. Students need to calculate and print double of each value (since two are connected in parallel).  
Print double of each capacitor value from a list.


```python
capacitance=[1*10^6,4*10^6,6*10^6,8*10^6,9*10^6,10*10^6]
print("Doubled capacitances are following:")
for c in capacitance:
    print(f"{c*2}uF")
```

## Task 7:  
A solar system records daily energy in kWh. Students must calculate and print half of each value (to simulate half-load usage).  
Print half of each solar energy reading from a list.


```python
solar_reading=[10,15,20,25,30]
for i,s in enumerate(solar_reading):
    print(f"day {i+1} half solar reading: {s/2}Kwh ")
```

## Task 1:
During charging, the battery slowly increases its percentage. Students want to simulate this charging process in code.  
Keep printing "Charging..." until battery percentage reaches 100%.  
	Start with battery = 0, increase by 10 in each loop, stop when 100.


```python
charging=0
while charging<100:
    print(f"{charging}% charging...")
    charging+=10
```

---
## Task 2:
Engineers use countdown timers for testing. Students must simulate a timer that runs before starting a test.
Countdown timer from 10 to 0, showing "Test Running".  
	Start variable = 10, decrease each loop, stop at 0


```python
start=10
print("test starting in:")
while start>=0:
    print(start)
    start-=1
```

## Task 3:
A voltage monitoring system asks for user input repeatedly until the system is turned off.  
Keep asking user for voltage until they enter 0.  
	Use input() in while loop, stop when entered value is 0.


```python
voltage=1
while voltage!=0:
    voltage=int(input("enter voltage please"))
```

## Task 4:
Power in a system doubles when load increases. Students need to simulate this doubling until it reaches a limit.  
Keep doubling the power until it exceeds 1000 W.  
	Start power = 1, multiply by 2 each loop, stop when >1000.


```python
power=1
while power<1000:
    print(power)
    power*=2
```

## Task 5:
A machine keeps running until its cycle count reaches a limit. Students must simulate this with a loop.  
Start from cycle = 1 and keep printing "Cycle Running" until cycle reaches 5.


```python
cycle=1
while cycle<5:
    print("cycle Running")
    cycle+=1
```

## Task 6:
A motor runs until its speed reaches 3000 RPM. Students must simulate this by starting from 0 and adding 500 each loop.  
Keep increasing speed by 500 until it reaches 3000, printing the speed each time.


```python
speed=0
while speed<3000:
    print("now speed:",speed)
    speed+=500
```
