## Task 4: Execute Code Only When No Error Occurs  
Problem:  
Display the calculated percentage and pass or fail status only if no exception occurs during input and validation.  
Use the Following Concepts  
	try  
	except  
	else


```python
try:
    total_marks=int(input("enter total marks"))
    obtained_marks=int(input("enter student obtained marks"))
except ValueError as error:
    print("Warraning: you entered wrong")
except ZeroDivisionError as error:
    print("you entered total marks 0")
if obtained_marks<0 or obtained_marks>total_marks:
    raise Exception("you entered wrong marks")
else:
    print("Student obtained ",(obtained_marks/total_marks)*100,"percent marks")
    if ((obtained_marks/total_marks)*100)>50:
        print("congratulation! ->pass")
    else:
        print("better luck next time ->fail")
```


## Task 5: Guarantee Program Completion Using finally  
Problem:  
No matter what happens during execution, the program must display:  
Result processing completed.  
This ensures proper program termination behavior.  
Use the Following Concepts  
	finally


```python
try:
    total_marks=int(input("enter total marks"))
    obtained_marks=int(input("enter student obtained marks"))
except ValueError as error:
    print("Warraning: you entered wrong")
except ZeroDivisionError as error:
    print("you entered total marks 0")
except:
  if obtained_marks<0 or obtained_marks>total_marks:
    raise Exception("you entered wrong marks")
else:
    print("Student obtained ",(obtained_marks/total_marks)*100,"percent marks")
    if ((obtained_marks/total_marks)*100)>50:
        print("congratulation! ->pass")
    else:
        print("better luck next time ->fail")
finally:
    print("Result processing completed")
```
