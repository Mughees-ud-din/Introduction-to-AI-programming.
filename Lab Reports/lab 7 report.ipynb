## You are working in a university electronics lab.   
You are asked to monitor a small solar home system installed on the lab roof. The system has:  
a)	A solar panel  
b)	A battery  
c)	A load controller  
d)	A temperature sensor  
Each minute, the system sends a small dictionary of readings  
1.	Create a dictionary named reading1 containing the 5 fields shown in the scenario.  
2.	From reading1, print:  
i.	The solar power  
ii.	The battery voltage  
iii.	The load status  
3.	Print how many fields the dictionary contains after your modifications.  
4.	Remove the "panel_current" field.  
5.	Get Keys, Values, and Items  
Print:  
i.	All keys  
ii.	All values  
iii.	All key-value pairs  
iv.	Use .keys(), .values(), and .items().


```python
#1
reading1={"solar_power":40,"battery_voltage":12,"load_status":"active","temp":35}

#2.1
print("solar power: ",reading1["solar_power"])
#2.2
print("battery voltage is: ",reading1["battery_voltage"])
#2.3
print("solar power: ",reading1["load_status"])
```

```python
#3
print(f"this dictionary have {len(reading1.keys())} fields")
```

```python
#4
del reading1["solar_power"]
print(reading1)
```

```python
#5
print("All keys are : ",reading1.keys())

print("All values are : ",reading1.values())

print("key value pairs are : ",reading1.items())
```
